# Merging and Concatenation

```python
import pandas as pd
```

```python
data={"Fellowship_ID":[1001,1002,1003,1004],
      "FirstName":["Olaoye","Ibukun","Abdul-Rauf","TItilayomi"],
      "Skill"  :["Praying","Coding","Programming","Speaking"]
     }
df=pd.DataFrame(data,index=list("ABCD"))
df
```

```python
data2={"Fellowship_ID":[1001,1005,1006,1007],
      "FirstName":["Olaoye","Kunle","Demola","Asake"],
      "Age"  :[40,49,62,51]
     }
df2=pd.DataFrame(data2,index=list("ABCD"))
df2
```

# Merge

```python
df.merge(df2,how="outer")
```

```python
df.merge(df2,how="inner")
```

```python
df.merge(df2,how="right")
```

```python
df.merge(df2,how="left")
```

```python
df.merge(df2,how = "cross")
```

# Concatenation
pd.concat([df,df2]) put on top of each other ...
pd.concat([df,df2],join="inner") takes only column that are similar in each tables

```python
pd.concat([df,df2])
```

```python
pd.concat([df,df2],join="inner")
```

```python
pd.concat([df,df2],join="outer")
```

```python
pd.concat(([df,df2]),join="outer", axis=1)
```

```python
pd.concat(([df,df2]),join="outer", axis=0)
```

# Append

```python
df.append(df2)
```

```python
df=pd.read_csv(r"C:\Users\HP\Documents\dirtydata.csv")
df

x= df["Calories"].median()
df["Calories"].fillna(x,inplace=True)
df
```

```python
df=pd.read_csv(r"C:\Users\HP\Documents\dirtydata.csv")
df

x= df["Calories"].mode()[0]
df["Calories"].fillna(x,inplace=True)
df
```
